```python
from collections import defaultdict
```

```python
def join(xs, ys, key='key', value='value', fields=[]):
    d=defaultdict(dict)
    for l, k in [(xs, 'x'), (ys,'y')]:
        for elem in l:
            d[elem[key]].update(
                {k:elem[value], 
                 key:elem[key]}
            )
            d[elem[key]].update(
                {a: b for a, b in elem.items() 
                 if a in fields}
            )
    return d.values()
```
